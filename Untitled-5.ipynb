```python
# ==========================================
# STEP 3: CORRECT MOVEMENT DETECTION LOGIC
# ==========================================

def detect_movements_correctly(data):
    """
    Correct movement detection based on your data structure:
    1. Each record shows what stage an account was in (from_stage)
    2. Movement = when consecutive records have different from_stages
    3. Same from_stage on multiple dates = no movement (just time spent in stage)
    """
    
    print("üîß STEP 3: BUILDING CORRECT MOVEMENT DETECTION")
    print("="*60)
    
    # Prepare data
    data = data.copy()
    data['from_stage_date'] = pd.to_datetime(data['from_stage_date'])
    
    all_movements = []
    
    # Process each techplay type
    for techplay in data['tech_play_type_label'].dropna().unique():
        
        print(f"üéØ Processing: {techplay}")
        
        techplay_data = data[data['tech_play_type_label'] == techplay]
        techplay_movements = 0
        accounts_processed = 0
        
        for account_id, account_data in techplay_data.groupby('oracle_account_id'):
            
            # Skip single-record accounts
            if len(account_data) < 2:
                continue
                
            accounts_processed += 1
            
            # Sort by date to get chronological order
            account_timeline = account_data.sort_values('from_stage_date')
            
            # Check each consecutive pair of records
            for i in range(1, len(account_timeline)):
                current_record = account_timeline.iloc[i]
                previous_record = account_timeline.iloc[i-1]
                
                current_stage = current_record['from_stage']
                previous_stage = previous_record['from_stage']
                
                # Movement detected ONLY if from_stage changed
                if current_stage != previous_stage:
                    techplay_movements += 1
                    
                    movement = {
                        'oracle_account_id': account_id,
                        'company_name': current_record['company_name'],
                        'techplay_type': techplay,
                        'from_stage': previous_stage,
                        'to_stage': current_stage,
                        'movement_date': current_record['from_stage_date'],
                        'previous_date': previous_record['from_stage_date'],
                        'days_between': (current_record['from_stage_date'] - previous_record['from_stage_date']).days,
                        'month_year': current_record['from_stage_date'].strftime('%Y-%m')
                    }
                    
                    all_movements.append(movement)
        
        print(f"   Accounts processed: {accounts_processed:,}")
        print(f"   Movements found: {techplay_movements:,}")
        
        if accounts_processed > 0:
            print(f"   Movement rate: {(techplay_movements/accounts_processed):.1f} movements per account")
    
    return pd.DataFrame(all_movements)

# Execute the correct detection
correct_movements = detect_movements_correctly(df)

print(f"\nüéØ CORRECT MOVEMENT RESULTS:")
print("="*60)
print(f"Total movements detected: {len(correct_movements):,}")

if len(correct_movements) > 0:
    print(f"Active accounts: {correct_movements['oracle_account_id'].nunique():,}")
    print(f"Active techplay types: {correct_movements['techplay_type'].nunique()}")
    
    print(f"\nüìä Top TechPlay Activity:")
    techplay_breakdown = correct_movements['techplay_type'].value_counts()
    for techplay, count in techplay_breakdown.head(10).items():
        print(f"   {techplay:<30}: {count:,} movements")
```

```python
# ==========================================
# F5 COMPLETE FUNNEL ACTIVITY ANALYSIS
# ==========================================

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.colors import LinearSegmentedColormap

def create_f5_funnel_activity_chart(data):
    """
    Create F5 funnel activity chart showing first-time entries into each stage by month
    Similar to the Takeout - F5/ADC chart you showed
    """
    
    print("üìä CREATING F5 COMPLETE FUNNEL ACTIVITY CHART")
    print("="*60)
    
    # Filter for F5 related techplays
    f5_techplays = ['Takeout - F5/ADC', 'F5', 'F5/ADC']  # Add any F5 variants in your data
    f5_data = data[data['tech_play_type_label'].isin(f5_techplays)].copy()
    
    if len(f5_data) == 0:
        print("‚ùå No F5 data found. Checking available F5 techplay types...")
        f5_related = data[data['tech_play_type_label'].str.contains('F5', case=False, na=False)]['tech_play_type_label'].unique()
        print(f"F5-related techplays found: {f5_related}")
        
        if len(f5_related) > 0:
            f5_data = data[data['tech_play_type_label'].isin(f5_related)].copy()
        else:
            return
    
    # Convert date
    f5_data['from_stage_date'] = pd.to_datetime(f5_data['from_stage_date'])
    
    # Define F5 funnel stages (customize based on your data)
    progression_stages = ['Qualification', 'Presentation', 'Proof of Concept', 'Initial Adoption', 'Mass Rollout']
    onhold_stages = ['On-Hold - Need Exec Engagement', 'On-Hold - Waiting on RFE', 'Customer - Deferred to SY26']
    rejection_stages = ['Rejected - Customer Rejection', 'Rejected - Solution Failure', 'Play does not Apply']
    
    all_stages = progression_stages + onhold_stages + rejection_stages
    
    # Find first-time entries into each stage for each account
    first_time_entries = []
    
    for account_id, account_data in f5_data.groupby('oracle_account_id'):
        account_timeline = account_data.sort_values('from_stage_date')
        stages_seen = set()
        
        for _, record in account_timeline.iterrows():
            stage = record['from_stage']
            
            # Only count first time in each stage
            if stage in all_stages and stage not in stages_seen:
                first_time_entries.append({
                    'oracle_account_id': account_id,
                    'company_name': record['company_name'],
                    'stage': stage,
                    'first_entry_date': record['from_stage_date'],
                    'month_year': record['from_stage_date'].strftime('%Y-%m')
                })
                stages_seen.add(stage)
    
    entries_df = pd.DataFrame(first_time_entries)
    
    if len(entries_df) == 0:
        print("‚ùå No stage entries found")
        return
    
    print(f"‚úÖ Found {len(entries_df)} first-time stage entries for F5")
    
    # Create monthly pivot
    monthly_entries = entries_df.pivot_table(
        index='month_year',
        columns='stage',
        values='oracle_account_id',
        aggfunc='count',
        fill_value=0
    )
    
    # Ensure all months are represented
    monthly_entries = monthly_entries.reindex(columns=all_stages, fill_value=0)
    
    # Create the stacked bar chart
    fig, ax = plt.subplots(figsize=(16, 10))
    
    # Define colors similar to your example
    colors = {
        # Progression stages (greens/blues)
        'Qualification': '#90EE90',
        'Presentation': '#2E8B57', 
        'Proof of Concept': '#4169E1',
        'Initial Adoption': '#FF6347',
        'Mass Rollout': '#DAA520',
        
        # On-hold stages (oranges)
        'On-Hold - Need Exec Engagement': '#FF8C00',
        'On-Hold - Waiting on RFE': '#FFA500',
        'Customer - Deferred to SY26': '#FFB84D',
        
        # Rejection stages (reds)
        'Rejected - Customer Rejection': '#DC143C',
        'Rejected - Solution Failure': '#B22222',
        'Play does not Apply': '#CD5C5C'
    }
    
    # Create stacked bars
    bottom = np.zeros(len(monthly_entries.index))
    
    for stage in all_stages:
        if stage in monthly_entries.columns and monthly_entries[stage].sum() > 0:
            values = monthly_entries[stage].values
            
            # Different styling for different stage types
            if stage in progression_stages:
                bars = ax.bar(range(len(monthly_entries.index)), values, bottom=bottom,
                             label=f'‚Üí {stage}', color=colors.get(stage, '#69b3a2'), 
                             alpha=0.8, edgecolor='white', linewidth=0.5)
            elif stage in onhold_stages:
                bars = ax.bar(range(len(monthly_entries.index)), values, bottom=bottom,
                             label=f'‚è∏ {stage}', color=colors.get(stage, '#ff7f0e'), 
                             alpha=0.8, edgecolor='black', linewidth=0.5)
            else:  # rejection stages
                bars = ax.bar(range(len(monthly_entries.index)), values, bottom=bottom,
                             label=f'‚ùå {stage}', color=colors.get(stage, '#d62728'), 
                             alpha=0.8, hatch='///', edgecolor='white', linewidth=0.5)
            
            # Add count labels on segments
            for i, (bar, value) in enumerate(zip(bars, values)):
                if value > 0:
                    label_y = bottom[i] + value/2
                    ax.text(bar.get_x() + bar.get_width()/2., label_y,
                           str(int(value)), ha='center', va='center', 
                           fontsize=10, fontweight='bold', 
                           color='white' if value > 3 else 'black')
            
            bottom += values
    
    # Formatting
    ax.set_title('F5 TechPlay: Complete Funnel Activity (WITH On-Hold)\n(First-time entries into each stage)', 
                fontsize=16, fontweight='bold', pad=20)
    ax.set_xlabel('Month', fontsize=14, fontweight='bold')
    ax.set_ylabel('Account Progressions', fontsize=14, fontweight='bold')
    
    # X-axis formatting
    month_labels = [date for date in monthly_entries.index]
    ax.set_xticks(range(len(month_labels)))
    ax.set_xticklabels(month_labels, rotation=45, ha='right')
    
    # Legend
    ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)
    
    # Grid
    ax.grid(True, alpha=0.3, axis='y')
    ax.set_axisbelow(True)
    
    plt.tight_layout()
    plt.show()
    
    # Print summary
    print(f"\nüìä F5 FUNNEL ACTIVITY SUMMARY:")
    print("="*50)
    
    stage_totals = monthly_entries.sum(axis=0).sort_values(ascending=False)
    total_entries = stage_totals.sum()
    
    print(f"Total first-time entries: {int(total_entries)}")
    print(f"Active months: {len(monthly_entries)}")
    print(f"Date range: {monthly_entries.index.min()} to {monthly_entries.index.max()}")
    
    print(f"\nStage breakdown:")
    for stage, count in stage_totals.items():
        if count > 0:
            percentage = (count / total_entries) * 100
            stage_type = "üéØ PROGRESSION" if stage in progression_stages else "‚è∏ ON-HOLD" if stage in onhold_stages else "‚ùå REJECTION"
            print(f"   {stage_type} {stage}: {int(count)} ({percentage:.1f}%)")
    
    # Monthly peaks
    monthly_totals = monthly_entries.sum(axis=1)
    peak_month = monthly_totals.idxmax()
    peak_count = monthly_totals.max()
    print(f"\nPeak activity month: {peak_month} ({int(peak_count)} entries)")
    
    return monthly_entries, entries_df

# ==========================================
# EXECUTE F5 FUNNEL ANALYSIS
# ==========================================

print("üöÄ CREATING F5 COMPLETE FUNNEL ACTIVITY ANALYSIS")
print("="*70)

# Create the F5 funnel chart
f5_monthly_entries, f5_entries_data = create_f5_funnel_activity_chart(df)

# Additional F5-specific analysis
if f5_entries_data is not None and len(f5_entries_data) > 0:
    
    print(f"\nüîç F5 FUNNEL INSIGHTS:")
    print("-"*50)
    
    # Conversion flow analysis
    progression_entries = f5_entries_data[f5_entries_data['stage'].isin(['Qualification', 'Presentation', 'Proof of Concept', 'Initial Adoption', 'Mass Rollout'])]
    
    if len(progression_entries) > 0:
        print(f"Accounts in progression stages: {progression_entries['oracle_account_id'].nunique()}")
        
        # Show progression funnel
        stage_counts = progression_entries['stage'].value_counts()
        print(f"\nF5 Progression Funnel:")
        funnel_stages = ['Qualification', 'Presentation', 'Proof of Concept', 'Initial Adoption', 'Mass Rollout']
        
        for i, stage in enumerate(funnel_stages):
            count = stage_counts.get(stage, 0)
            if i == 0:
                print(f"   {stage}: {count} accounts (100%)")
                base_count = max(count, 1)
            else:
                percentage = (count / base_count) * 100 if base_count > 0 else 0
                print(f"   {stage}: {count} accounts ({percentage:.1f}%)")
    
    # Account journey complexity
    account_stage_counts = f5_entries_data.groupby('oracle_account_id')['stage'].count()
    complex_accounts = account_stage_counts[account_stage_counts >= 3]
    
    if len(complex_accounts) > 0:
        print(f"\nComplex F5 journeys (3+ stages): {len(complex_accounts)} accounts")
        most_complex = account_stage_counts.idxmax()
        max_stages = account_stage_counts.max()
        
        complex_company = f5_entries_data[f5_entries_data['oracle_account_id'] == most_complex]['company_name'].iloc[0]
        print(f"Most complex journey: {complex_company} ({max_stages} stages)")

print(f"\n‚úÖ F5 FUNNEL ANALYSIS COMPLETE!")
```

```python
# ==========================================
# F5 ACTUAL TRANSITIONS (FIXED VERSION)
# ==========================================

def analyze_f5_actual_transitions_fixed(data):
    """
    Fixed version - convert dates to datetime first
    """
    
    print("üéØ F5 ACTUAL TRANSITIONS ANALYSIS (FIXED)")
    print("="*60)
    
    # Filter for F5 data
    f5_related = data[data['tech_play_type_label'].str.contains('F5', case=False, na=False)].copy()
    
    if len(f5_related) == 0:
        print("‚ùå No F5 data found")
        return None
    
    # FIX: Convert date column to datetime
    f5_related['from_stage_date'] = pd.to_datetime(f5_related['from_stage_date'])
    
    print(f"üìä F5 Dataset: {len(f5_related):,} records, {f5_related['oracle_account_id'].nunique():,} accounts")
    
    # Apply movement detection logic
    f5_movements = []
    
    for account_id, account_data in f5_related.groupby('oracle_account_id'):
        if len(account_data) < 2:
            continue
            
        # Sort chronologically
        account_timeline = account_data.sort_values('from_stage_date')
        
        # Compare consecutive records for stage changes
        for i in range(1, len(account_timeline)):
            current_record = account_timeline.iloc[i]
            previous_record = account_timeline.iloc[i-1]
            
            current_stage = current_record['from_stage']
            previous_stage = previous_record['from_stage']
            
            # Movement = stage actually changed
            if current_stage != previous_stage:
                f5_movements.append({
                    'oracle_account_id': account_id,
                    'company_name': current_record['company_name'],
                    'from_stage': previous_stage,
                    'to_stage': current_stage,
                    'movement_date': current_record['from_stage_date'],
                    'days_between': (current_record['from_stage_date'] - previous_record['from_stage_date']).days,
                    'month_year': current_record['from_stage_date'].strftime('%Y-%m')
                })
    
    movements_df = pd.DataFrame(f5_movements)
    
    print(f"‚úÖ ACTUAL F5 TRANSITIONS FOUND: {len(movements_df)}")
    print(f"‚úÖ Accounts with transitions: {movements_df['oracle_account_id'].nunique() if len(movements_df) > 0 else 0}")
    
    if len(movements_df) == 0:
        print("‚ùå No F5 transitions found")
        return movements_df
    
    # ==========================================
    # TRANSITION ANALYSIS
    # ==========================================
    
    print(f"\nüìã F5 STAGE TRANSITION BREAKDOWN:")
    print("-"*70)
    
    # Most common transitions
    transition_paths = movements_df.apply(lambda x: f"{x['from_stage']} ‚Üí {x['to_stage']}", axis=1)
    top_transitions = transition_paths.value_counts()
    
    total_transitions = len(movements_df)
    
    for i, (transition, count) in enumerate(top_transitions.head(10).items()):
        percentage = (count / total_transitions) * 100
        print(f"{i+1:2d}. {transition:<60} {count:2d} ({percentage:4.1f}%)")
    
    # ==========================================
    # MONTHLY TRANSITION ACTIVITY
    # ==========================================
    
    print(f"\nüìÖ F5 MONTHLY TRANSITION ACTIVITY:")
    print("-"*50)
    
    monthly_transitions = movements_df['month_year'].value_counts().sort_index()
    
    for month, count in monthly_transitions.items():
        percentage = (count / total_transitions) * 100
        print(f"   {month}: {count:2d} transitions ({percentage:4.1f}%)")
    
    # ==========================================
    # SAMPLE TRANSITIONS
    # ==========================================
    
    print(f"\nüè¢ SAMPLE F5 TRANSITIONS:")
    print("-"*80)
    
    sample_transitions = movements_df.head(10)
    for i, (_, movement) in enumerate(sample_transitions.iterrows()):
        print(f"{i+1:2d}. {movement['company_name'][:30]:<30} | "
              f"{movement['movement_date'].strftime('%Y-%m-%d')} | "
              f"{movement['from_stage']:<25} ‚Üí {movement['to_stage']:<25} | "
              f"({movement['days_between']} days)")
    
    # ==========================================
    # SUMMARY STATS
    # ==========================================
    
    print(f"\nüìä F5 TRANSITION SUMMARY:")
    print("="*60)
    print(f"Total F5 transitions: {len(movements_df)}")
    print(f"Active F5 accounts: {movements_df['oracle_account_id'].nunique()}")
    print(f"Average days between transitions: {movements_df['days_between'].mean():.1f}")
    print(f"Date range: {movements_df['movement_date'].min().strftime('%Y-%m-%d')} to {movements_df['movement_date'].max().strftime('%Y-%m-%d')}")
    
    # Most active month
    if len(monthly_transitions) > 0:
        best_month = monthly_transitions.idxmax()
        best_count = monthly_transitions.max()
        print(f"Most active month: {best_month} ({best_count} transitions)")
    
    return movements_df

# ==========================================
# EXECUTE FIXED VERSION
# ==========================================

# Run the fixed analysis
f5_actual_transitions = analyze_f5_actual_transitions_fixed(df)

# ==========================================
# SIMPLE VISUALIZATION
# ==========================================

if f5_actual_transitions is not None and len(f5_actual_transitions) > 0:
    
    print(f"\nüìä Creating F5 Transitions Chart...")
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
    
    # Chart 1: Top Transition Paths
    transition_paths = f5_actual_transitions.apply(lambda x: f"{x['from_stage'][:20]} ‚Üí {x['to_stage'][:20]}", axis=1)
    top_paths = transition_paths.value_counts().head(8)
    
    bars = ax1.barh(range(len(top_paths)), top_paths.values, color='steelblue', alpha=0.8)
    ax1.set_title(f'F5 Top Transition Paths\n({len(f5_actual_transitions)} Total Transitions)', fontweight='bold')
    ax1.set_xlabel('Number of Transitions')
    ax1.set_yticks(range(len(top_paths)))
    ax1.set_yticklabels(top_paths.index, fontsize=9)
    
    for i, bar in enumerate(bars):
        width = bar.get_width()
        ax1.text(width + 0.1, bar.get_y() + bar.get_height()/2.,
                f'{int(width)}', ha='left', va='center', fontweight='bold')
    
    # Chart 2: Monthly Activity
    monthly_activity = f5_actual_transitions['month_year'].value_counts().sort_index()
    
    bars = ax2.bar(range(len(monthly_activity)), monthly_activity.values, 
                   color='forestgreen', alpha=0.8, edgecolor='darkgreen')
    ax2.set_title('F5 Monthly Transitions', fontweight='bold')
    ax2.set_ylabel('Number of Transitions')
    ax2.set_xlabel('Month')
    ax2.set_xticks(range(len(monthly_activity)))
    ax2.set_xticklabels(monthly_activity.index, rotation=45)
    
    for bar in bars:
        height = bar.get_height()
        ax2.text(bar.get_x() + bar.get_width()/2., height + 0.1,
                f'{int(height)}', ha='center', va='bottom', fontweight='bold')
    
    plt.tight_layout()
    plt.show()
    
    print(f"\n‚úÖ F5 TRANSITIONS ANALYSIS COMPLETE!")
    
else:
    print(f"\n‚ùå No F5 transitions to analyze")

# Quick comparison
print(f"\nüîç F5 TRANSITIONS vs ENTRIES COMPARISON:")
print(f"F5 Funnel Entries: 2,123 (first-time stage appearances)")
print(f"F5 Actual Transitions: {len(f5_actual_transitions) if f5_actual_transitions is not None else 0} (stage-to-stage movements)")
```
