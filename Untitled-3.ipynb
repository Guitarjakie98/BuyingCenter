```python
import pandas as pd
from fuzzywuzzy import fuzz, process
import numpy as np

# First, let's examine the data structure to make sure we're working with the right columns
print("Accounts columns:", Accounts.columns.tolist())
print("DatabaseAccounts columns:", DatabaseAccounts.columns.tolist())

# Function to find the best fuzzy match
def find_best_match(name, database_names, threshold=80):
    """
    Find the best fuzzy match for a name in the database
    Returns the best match and its score, or None if below threshold
    """
    if pd.isna(name) or name == '':
        return None, 0
    
    # Get the best match using fuzzy matching
    best_match = process.extractOne(name, database_names, scorer=fuzz.ratio)
    
    if best_match and best_match[1] >= threshold:
        return best_match[0], best_match[1]
    else:
        return None, best_match[1] if best_match else 0

# Create a copy of Accounts to work with
result_df = Accounts.copy()

# Get unique customer names from DatabaseAccounts (assuming the column is 'Oracle Account Customer Name')
# Adjust this column name based on your actual data
customer_name_column = 'Oracle Account Customer Name'  # Update this if different

# Remove any null/empty names from the database for cleaner matching
database_names = DatabaseAccounts[customer_name_column].dropna().unique().tolist()
database_names = [name for name in database_names if str(name).strip() != '']

print(f"Found {len(database_names)} unique customer names in database")

# Initialize new columns for the results
result_df['Match_Found'] = 'No'
result_df['Match_Score'] = 0
result_df['Matched_Customer_Name'] = ''

# Add columns from DatabaseAccounts that you want to pull in
# Adjust these column names based on what you want to include
columns_to_add = [
    'Oracle Account Customer ID',
    'Oracle Account Account Segmentation', 
    'Oracle Account Business Unit',
    'AE Person Name',
    'AE Level14 Territory Name'
    # Add more columns as needed
]

# Initialize these columns in the result DataFrame
for col in columns_to_add:
    if col in DatabaseAccounts.columns:
        result_df[col] = ''

# Perform fuzzy matching for each account
print("Performing fuzzy matching...")
for index, row in result_df.iterrows():
    account_name = row['Name']
    
    # Find best match
    best_match, score = find_best_match(account_name, database_names, threshold=80)
    
    if best_match:
        # Update match information
        result_df.at[index, 'Match_Found'] = 'Yes'
        result_df.at[index, 'Match_Score'] = score
        result_df.at[index, 'Matched_Customer_Name'] = best_match
        
        # Get the corresponding row from DatabaseAccounts
        matched_row = DatabaseAccounts[DatabaseAccounts[customer_name_column] == best_match].iloc[0]
        
        # Copy over the specified columns
        for col in columns_to_add:
            if col in DatabaseAccounts.columns:
                result_df.at[index, col] = matched_row[col]
        
        print(f"✓ Matched '{account_name}' with '{best_match}' (Score: {score})")
    else:
        print(f"✗ No match found for '{account_name}' (Best score: {score})")

# Display results summary
print(f"\nMatching Summary:")
print(f"Total accounts: {len(result_df)}")
print(f"Matches found: {len(result_df[result_df['Match_Found'] == 'Yes'])}")
print(f"No matches: {len(result_df[result_df['Match_Found'] == 'No'])}")

# Show the results
print(f"\nFirst few results:")
display_columns = ['Name', 'Match_Found', 'Match_Score', 'Matched_Customer_Name'] + [col for col in columns_to_add if col in result_df.columns]
print(result_df[display_columns].head(10))

# Optional: Save results to CSV
result_df.to_csv('matched_accounts_results.csv', index=False)
print(f"\nResults saved to 'matched_accounts_results.csv'")
```
