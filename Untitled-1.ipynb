```python
import pandas as pd

Eventt = pd.read_csv(
    '/Applications/WorkDataSets/DataStore/1HData.csv',
    encoding='latin1'  # or 'ISO-8859-1'
)

print(Eventt.info())
```

```python
EloquaContactDB = pd.read_csv('/Applications/WorkDataSets/CDP Project/Clean CDP Data 2.0/EloquaContactDatabase.csv', encoding='utf-8')
print(EloquaContactDB.info())
```

```python
import re
import pandas as pd

# --- CONFIG ---
# If your Eloqua DB file is already loaded as a DataFrame, skip the read step below.
# Otherwise, point this to your Eloqua contacts export:
# EloquaContactDB = pd.read_csv('/path/to/eloqua_contacts.csv', encoding='latin1')

# Column name candidates (in case headers vary between exports)
EVENT_EMAIL_COL_CANDIDATES  = ['email address', 'email', 'email_address']
ELOQUA_EMAIL_COL_CANDIDATES = ['email address', 'email', 'email_address']

def find_col(df, candidates):
    cols = [c for c in df.columns.str.lower().str.strip()]
    mapping = {c.lower().strip(): c for c in df.columns}
    for cand in candidates:
        if cand in cols:
            return mapping[cand]
    raise KeyError(f"None of {candidates} found in columns: {list(df.columns)}")

# 1) Normalize headers
Eventt.columns = Eventt.columns.str.strip()
EloquaContactDB.columns = EloquaContactDB.columns.str.strip()

# 2) Locate the email columns
event_email_col  = find_col(Eventt, EVENT_EMAIL_COL_CANDIDATES)
eloqua_email_col = find_col(EloquaContactDB, ELOQUA_EMAIL_COL_CANDIDATES)

# 3) Clean & validate emails
EMAIL_RE = re.compile(r"^[A-Z0-9._%+-]+@[A-Z0-9.-]+\.[A-Z]{2,}$", re.IGNORECASE)

def clean_email_series(s: pd.Series) -> pd.Series:
    s = s.astype(str).str.strip().str.lower()
    s = s.where(s.ne('nan'))  # turn literal "nan" strings into real NaN
    s = s.where(s.ne('none'))
    s = s.where(s.ne(''))     # empty -> NaN
    return s

Eventt[event_email_col]  = clean_email_series(Eventt[event_email_col])
EloquaContactDB[eloqua_email_col] = clean_email_series(EloquaContactDB[eloqua_email_col])

Eventt['__email_valid__']  = Eventt[event_email_col].apply(lambda x: bool(EMAIL_RE.match(x)) if pd.notna(x) else False)
EloquaContactDB['__email_valid__'] = EloquaContactDB[eloqua_email_col].apply(lambda x: bool(EMAIL_RE.match(x)) if pd.notna(x) else False)

# Optional: drop obviously invalid addresses from either side
Eventt_valid          = Eventt.loc[Eventt['__email_valid__']].copy()
EloquaContactDB_valid = EloquaContactDB.loc[EloquaContactDB['__email_valid__']].copy()

# 4) De-duplicate by email (keep first occurrence)
Eventt_valid = Eventt_valid.drop_duplicates(subset=[event_email_col])
EloquaContactDB_valid = EloquaContactDB_valid.drop_duplicates(subset=[eloqua_email_col])

# 5) Build set for fast membership test
db_emails = set(EloquaContactDB_valid[eloqua_email_col].dropna().tolist())

# 6) Filter IN (present in Eloqua) and OUT (not present in Eloqua)
Cross_event_in_db  = Eventt_valid[Eventt_valid[event_email_col].isin(db_emails)].copy()
Cross_event_not_in = Eventt_valid[~Eventt_valid[event_email_col].isin(db_emails)].copy()

# 7) Quick summary
total_raw   = len(Eventt)
total_valid = len(Eventt_valid)
in_ct       = len(Cross_event_in_db)
out_ct      = len(Cross_event_not_in)

print("=== Email Filter Summary ===")
print(f"Event rows (raw):                 {total_raw:,}")
print(f"Event rows with valid emails:     {total_valid:,}")
print(f"In Eloqua (kept):                 {in_ct:,}")
print(f"Not in Eloqua (removed):          {out_ct:,}")
print(f"Retention rate (valid -> kept):   {in_ct/total_valid:.1%}" if total_valid else "Retention rate: n/a")

# 8) (Optional) save results
# Cross_event_in_db.to_csv('/Applications/WorkDataSets/DataStore/CCROSSEVENT_filtered_in.csv', index=False)
# Cross_event_not_in.to_csv('/Applications/WorkDataSets/DataStore/CCROSSEVENT_filtered_out.csv', index=False)

# 9) If you still need the filtered view with the original (non-deduped) Eventt:
# Map Eloqua membership back to the original Eventt
Eventt['__in_eloqua__'] = Eventt[event_email_col].isin(db_emails)
Cross_event_filtered_full = Eventt[Eventt['__in_eloqua__']].copy()
```
