```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt

df = pd.read_csv('/Applications/WorkDataSets/DataStore/Demandbase CDP 4.0/DemandbaseDataCDP2.csv')

df.info()
```

```python
import pandas as pd
import re

# List of company names to search for
company_names = [
    'Best Buy',
    'Jeffries',
    'Pepsi',
    'M&T Bank',
    'State Street',
    'Humana',
    'Booz Allen',
    'USAA',
    'Synchrony',
    'Textron',
    'Credit Agricole',
    'BNP Paribas',
    'Soc Gen',
    'Siemens',
    'Ericsson',
    'Fiserv'
]

# Create a pattern that matches any of these company names (case-insensitive)
pattern = '|'.join([re.escape(name) for name in company_names])

# Search in the Account Name column
matches = df[df['Account Name'].str.contains(pattern, case=False, na=False)]

print(f"Found {len(matches)} matches:")
print(matches[['Account Name', 'First Name', 'Last Name', 'Job Title']])

# If you want to see which specific companies were found:
found_companies = []
for name in company_names:
    company_matches = df[df['Account Name'].str.contains(name, case=False, na=False)]
    if len(company_matches) > 0:
        found_companies.append((name, len(company_matches)))

print("\nCompany match summary:")
for company, count in found_companies:
    print(f"{company}: {count} matches")
```

```python
import pandas as pd
import re
import os

# Create directory if it doesn't exist
output_dir = '/Applications/WorkDataSets/DataStore/F5OCT/'
os.makedirs(output_dir, exist_ok=True)

# Function to clean filename
def clean_filename(name):
    # Remove special characters and replace spaces with underscores
    cleaned = re.sub(r'[^\w\s-]', '', name)
    cleaned = re.sub(r'[-\s]+', '_', cleaned)
    return cleaned.strip('_')

# Get unique account names from your matches
unique_accounts = matches['Account Name'].unique()

print(f"Exporting data for {len(unique_accounts)} unique accounts...")

for account_name in unique_accounts:
    # Filter all activities for this account
    account_data = df[df['Account Name'] == account_name]
    
    if len(account_data) > 0:
        # Create clean filename
        clean_name = clean_filename(account_name)
        filename = f"{clean_name}_activities.csv"
        filepath = os.path.join(output_dir, filename)
        
        # Save to CSV
        account_data.to_csv(filepath, index=False)
        print(f"Exported {len(account_data)} records for '{account_name}' to {filename}")

print("Export complete!")
```

```python
import pandas as pd
import re
import os

# Create directory if it doesn't exist
output_dir = '/Applications/WorkDataSets/DataStore/F5OCT/'
os.makedirs(output_dir, exist_ok=True)

def clean_filename(name):
    cleaned = re.sub(r'[^\w\s-]', '', name)
    cleaned = re.sub(r'[-\s]+', '_', cleaned)
    return cleaned.strip('_')

# Get unique account names from your matches
unique_accounts = matches['Account Name'].unique()

print(f"Exporting data with names for {len(unique_accounts)} unique accounts...")

for account_name in unique_accounts:
    # Filter all activities for this account
    account_data = df[df['Account Name'] == account_name]
    
    # Filter to only include rows with actual names (not NaN)
    account_with_names = account_data[
        (account_data['First Name'].notna()) & 
        (account_data['Last Name'].notna()) &
        (account_data['First Name'] != '') & 
        (account_data['Last Name'] != '')
    ]
    
    if len(account_with_names) > 0:
        # Create clean filename
        clean_name = clean_filename(account_name)
        filename = f"{clean_name}_contacts.csv"
        filepath = os.path.join(output_dir, filename)
        
        # Save to CSV
        account_with_names.to_csv(filepath, index=False)
        print(f"Exported {len(account_with_names)} records with names for '{account_name}' to {filename}")
        print(f"  (Filtered from {len(account_data)} total records)")
    else:
        print(f"No records with names found for '{account_name}'")

print("\nExport complete!")
```

```python
import pandas as pd
import re
import os

# Create directory if it doesn't exist
output_dir = '/Applications/WorkDataSets/DataStore/F5OCT/'
os.makedirs(output_dir, exist_ok=True)

# Get all the matched company data and filter for records with names
all_company_contacts = matches[
    (matches['First Name'].notna()) & 
    (matches['Last Name'].notna()) &
    (matches['First Name'] != '') & 
    (matches['Last Name'] != '')
]

# Sort by Account Name for better organization
all_company_contacts = all_company_contacts.sort_values(['Account Name', 'Last Name', 'First Name'])

# Save to one combined CSV
filename = 'all_target_companies_contacts.csv'
filepath = os.path.join(output_dir, filename)
all_company_contacts.to_csv(filepath, index=False)

print(f"Exported {len(all_company_contacts)} total contact records to {filename}")
print(f"Original matches: {len(matches)} records")
print(f"Records with names: {len(all_company_contacts)} records")

# Show summary by company
print("\nBreakdown by company:")
company_summary = all_company_contacts['Account Name'].value_counts()
for account, count in company_summary.head(10).items():
    print(f"  {account}: {count} contacts")

if len(company_summary) > 10:
    print(f"  ... and {len(company_summary) - 10} more companies")

print(f"\nFile saved to: {filepath}")
```
