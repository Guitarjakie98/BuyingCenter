```python
import pandas as pd

# --- CONFIG ---
INPUT_PATH = "/Applications/WorkDataSets/combined_cleaned_full.csv"
OUTPUT_PATH = "/Applications/WorkDataSets/combined_cleaned_full_CLEANED.csv"

# List of partner account IDs to remove
partner_account_ids = [
    "123456", "987654", "555555"  # <-- replace these with real account IDs
]

# --- LOAD DATA ---
print("Loading dataset...")
df = pd.read_csv(INPUT_PATH)
print(f"Loaded {len(df):,} rows and {len(df.columns)} columns")

# --- CHECK COLUMN ---
# Use either 'Account Number' or 'Account Name' depending on your dataset
account_col = "Account Number"
if account_col not in df.columns:
    raise ValueError(f"Column '{account_col}' not found. Available columns:\n{list(df.columns)}")

# --- FILTER OUT PARTNER ACCOUNTS ---
before = len(df)
df = df[~df[account_col].isin(partner_account_ids)]
after = len(df)
removed = before - after
print(f"Removed {removed:,} partner account rows ({removed / before:.2%} of dataset)")

# --- SAVE CLEANED FILE ---
df.to_csv(OUTPUT_PATH, index=False)
print(f"âœ… Cleaned file saved to: {OUTPUT_PATH}")
```
